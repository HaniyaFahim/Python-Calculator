```python
import tkinter as tk
from tkinter import *
```

```python
root = tk.Tk()
root.geometry("430x540")
root.configure(bg="#333")
root.title("Calculator")

operator = ""

text_input = StringVar()

def btnClick(number):
    global operator
    operator = operator + str(number)
    text_input.set(operator)

    
def btnEquals():
    global operator
    ans = str(eval(operator))
    text_input.set(ans)
    
    
def btnClear():
    global operator
    operator = ""
    text_input.set("")


e1 = Entry(root, textvariable=text_input, justify='right', font=(  'Segoe Print',20,'bold'),bg="orange",fg="black" , borderwidth=12, relief="ridge")
e1.grid(row=0,columnspan=4,rowspan=2)

b1=Button(root,text='7', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" , borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(7))
b1.grid(row=2,column=0)

b2=Button(root,text='8', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(8))
b2.grid(row=2,column=1)

b3=Button(root,text='9', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(9))
b3.grid(row=2,column=2)

b4=Button(root,text='+', font=( 'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove", height=2,width=8 ,command=lambda:btnClick("+"))
b4.grid(row=2,column=3)


b5=Button(root,text='4', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" , borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(4))
b5.grid(row=3,column=0)

b6=Button(root,text='5', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(5))
b6.grid(row=3,column=1)

b7=Button(root,text='6', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(6))
b7.grid(row=3,column=2)

b8=Button(root,text='-', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick("-"))
b8.grid(row=3,column=3)


b9=Button(root,text='1', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" , borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(1))
b9.grid(row=4,column=0)

b10=Button(root,text='2', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(2))
b10.grid(row=4,column=1)

b11=Button(root,text='3', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(3))
b11.grid(row=4,column=2)

b12=Button(root,text='*', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick("*"))
b12.grid(row=4,column=3)


b13=Button(root,text='0', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" , borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick(0))
b13.grid(row=5,column=0)

b14=Button(root,text='Clear', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=18,command= btnClear)
b14.grid(row=6,column=0, columnspan=2)

b16=Button(root,text='/', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick("/"))
b16.grid(row=5,column=2)

b15=Button(root,text='=', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=5,width=8,command=btnEquals)
b15.grid(row=5,column=3,rowspan=2)




b14=Button(root,text='.', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick("."))
b14.grid(row=5,column=1)

b15=Button(root,text='%', font=(  'Segoe Print',12,'bold'),bg="black",fg="orange" ,bd=12, borderwidth=10, relief="groove",height=2,width=8,command=lambda:btnClick("%"))
b15.grid(row=6,column=2)




root.mainloop()
```
